```python
import pandas as pd, numpy as np, requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import re
```

## Final code

```python
review=[]
review_title=[]
rating=[]
loc=[]
deg=[]
date=[]
for i in range(0,7001,20):
    url='https://in.indeed.com/cmp/Starbucks/reviews?fcountry=ALL&start='+str(i)+'&lang=en'
    driver = webdriver.Chrome(r"C:\Users\sreya\Downloads\chromedriver_win32\chromedriver.exe")
    driver.get(url)
    #print(url)
    soup = bs(driver.page_source,'html.parser')
    driver.close()
    
    container = soup.find_all('div',class_='css-t3vkys eu4oa1w0')
    
    #for review
    for i in container:
        review_cls = i.find('span',{'class' :'css-1cxc9zk e1wnkr790'})
        
        for y in review_cls:
            a=y.text
            if a =='':
                continue
            else:
                review.append(a)
                
    #for review title
        reviewtitle_cls = i.find('a',class_='css-5ytf38 emf9s7v0')
        for r in reviewtitle_cls:
            review_title.append(r.text)
            
    # for rating   
        rating_cls = i.find('button',{'class' :'css-1c33izo e1wnkr790'})
        for x in rating_cls:
            rating.append(x.text)
            
    # for location
    for i in container:
        loc_cls = i.find('div',{'class' :'css-8a5o2x e1wnkr790'})
        for z in loc_cls:
            x = z.text
            reg = re.findall('\-\s(.+)\s\-',x)
            loc.append(reg[0])
            
    # for designation
    for i in container:
        loc_cls = i.find('div',{'class' :'css-8a5o2x e1wnkr790'})
        for z in loc_cls:
            x = z.text
            reg = re.findall('(.*)\s\(',x)
            deg.append(reg[0])
            
    # for Date
    for i in container:
        loc_cls = i.find('div',{'class' :'css-8a5o2x e1wnkr790'})
        for z in loc_cls:
            x = z.text
            reg = re.findall('\s\-\s(\w+\s\d+\,\s\d+)',x)
            date.append(reg[0])
    
```

```python
df = pd.DataFrame({'Location':loc,'date':date,'Designation':deg,'Title':review_title,'Review':review,"Rating":rating})
df
```

```python
# REmoving the duplicates

df1 = df.drop_duplicates(ignore_index=True)
```

```python
#Changing the datatypes

df1['date'] = df1['date'].astype('datetime64[ns]')
```

```python
df1['Rating'] = df1['Rating'].astype('float')
```
